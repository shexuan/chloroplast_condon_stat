```python
import os 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
data_dir = ''
os.chdir(data_dir)
os.getcwd()
```

```python
enc = lambda x: (2 + x + 29 / (x**2 + (1 - x)**2))
fig, ax = plt.subplots(2,3,squeeze=False,figsize=(20,15))
def plot_enc(file,tit,axes):
    #plt.subplot(231)
    df = pd.read_excel(file, header=0)
    #df
    # 
    x = list(np.linspace(0,1,201))
    y = [enc(i) for i in x]
    axes.plot(x,y)
    axes.scatter(df['GC3s'], df['Nc'])
    axes.title.set_text(tit)
    axes.set_xlabel('GC3s')
    axes.set_ylabel('ENC')
    
plt.figure(dpi=300,figsize=(20,15))
plot_enc('1lt.rscu.xlsx', '1LT', ax[0,0])
plot_enc('2ts.rscu.xlsx', '2TS', ax[0,1])
plot_enc('3ms.rscu.xlsx', '3MS', ax[0,2])
plot_enc('4mf.rscu.xlsx', '4MF', ax[1,0])
plot_enc('5mg.rscu.xlsx', '5MG', ax[1,1])
plt.savefig('enc_plot.png')
```
